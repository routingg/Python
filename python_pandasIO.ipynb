```python
'''
# UC Irvine Machine Learning Repository에서 제공하는 자동차 연비 데이터셋
# 출처 :  https://archive.ics.uci.edu/dataset/9/auto+mpg

    속성(feature)                         데이터 타입 - 범위(상세)

    1. mpg:             연비             continuous                   연속값
    2. cylinders:       실린더수         multi-valued discrete        이산값
    3. displacement:    배기량           continuous
    4. horsepower:      마력(출력)       continuous
    5. weight:          차중(무게)       continuous
    6. acceleration:    가속능력         continuous
    7. model year:      출시년도         multi-valued discrete
    8. origin:          제조국           multi-valued discrete
    9. car name:        자동차이름       string (unique for each instance)

'''
```

```python
# 데이터 파일 경로 확인 후 데이터 읽어들이기
# csv to df

df = pd.read_csv('./auto-mpg.csv')
```

```python
# 파일을 읽어들인 후 전체를 확인하는 것은 권장사항이 아님 (데이터가 매우 클 경우)
# 일부만 확인하도록 하는 방법이 권장

df.head()       #head() :  위에서부터 5개 행만 출력
```

```python
# 위에서부터 5개 행 확인 결과 - 첫번재 행이 컬럼으로 잘못 지정
df = pd.read_csv('./auto-mpg.csv', header=None)
df.head()
```

```python
# 컬럼 이름을 지정
df.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration',
             'model year', 'origin', 'car name']
df.head()
```

```python
# 원본데이터를 처리한 기본데이터 저장
df.to_csv('./auto-mpg-header.csv')
```

### 판다스 내장 그래프 도구

```python
df = pd.read_excel('./남북한발전전력량.xlsx')
df.head()
```

```python
# 남한, 북한 발전전력량 중 합계 분만 따로 추출
df_south_north = df.iloc[[0, 5], 2:]
df_south_north
```

```python
df_south_north.info()
```

```python
# 기술 통계
df_south_north.describe()
```

```python
# 행 인덱스 이름 변경
df_south_north.index = ['South', 'North']
df_south_north
```

```python
# 컬럼 인덱스 객체 자료형을 숫자형으로 변경
df_south_north.columns = df_south_north.columns.map(int)

print(type(df_south_north.columns))
```

```python
# 선 그래프 : plot()

df_south_north.plot()
```

```python
# 행, 열 전치
df_sn_t = df_south_north.T
df_sn_t.head()
```

```python
df_sn_t.plot()
```

```python
# 막대 그래프
```

```python
df_sn_t.plot(kind='bar')
```
